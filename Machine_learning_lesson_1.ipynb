### **Array Common Manipulation**

manual construction of array

```python
#creating of single and multi array
a1 = np.array([2,4,5,7,11])
a1i = np.array([3,6,10,13,11])
a2= np.array([[1,4,8,9],[1,3,5,7]])
a2i= np.array([[2,4,7,5],[1,3,4,5]])


a1,a2,a1.shape,a2.shape
```

```python
#use of range function to evenly distribute numbers by increment and decrement
b1= np.arange(10)
b2= np.arange(11,20)
b3= np.arange(21,30,2)

b1,b2,b3
```

```python

#another use of increment and decrement by point
c1= np.linspace(31,32,3)
c2= np.linspace(35,40,20)


c1,c2
```

```python
#creating same types of array
e1 = np.ones((3,5))
e2= 7 * np.ones((3,3))
e3 = 0* np.ones((3,5))

e1,e2,e3
```
